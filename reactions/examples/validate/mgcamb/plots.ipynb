```python
import sympy as sp
from sympy import *
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
from scipy.integrate import quad
from scipy.optimize import fmin
import scipy.integrate as integrate
import scipy.special as special
import scipy.stats as st
import sys
font1 = {'size'   : 20, 'family':'STIXGeneral'}

from platform import python_version

print(python_version())
```

```python
#Linear spectrum comparisons 

fig0 = plt.figure()
plt.figure(figsize=(10,5))

dat = np.loadtxt("linear/mnu_0.4_mmpower_z0.dat")
dat2 = np.loadtxt("linear/mnu_0.4_mmpower_z1.dat")

datb = np.loadtxt("linear/mmpower_z0.dat")
dat2b = np.loadtxt("linear/mmpower_z1.dat")

ax = plt.subplot(1, 3, 1)
plt.xscale('log')
plt.plot(dat[:,0], dat[:,1]/datb[:,1] ,linestyle='dashed', color='b', label= 'z=0',alpha=1.)
plt.plot(dat2[:,0], dat2[:,1]/dat2b[:,1] ,linestyle='dashed', color='b', label= 'z=0',alpha=1.)
ax.tick_params(axis='both', which='major', labelsize=15)
plt.ylim(0.9,1.1)
plt.xlim(0.001,30)
plt.ylabel('$P_{m}/P_{b}$', **font1)
plt.xlabel('k[h/Mpc]',**font1)
plt.legend(loc='best',prop={'size': 14})
plt.tight_layout(pad=3.0)

dat = np.loadtxt("linear/mnu_0.4_cbpower_z0.dat")
dat2 = np.loadtxt("linear/mnu_0.4_cbpower_z1.dat")

datb = np.loadtxt("linear/cbpower_z0.dat")
dat2b = np.loadtxt("linear/cbpower_z1.dat")

ax = plt.subplot(1, 3, 2)
plt.xscale('log')
plt.plot(dat[:,0], dat[:,1]/datb[:,1] ,linestyle='dashed', color='b', label= 'z=0',alpha=1.)
plt.plot(dat2[:,0], dat2[:,1]/dat2b[:,1] ,linestyle='dashed', color='b', label= 'z=0',alpha=1.)
ax.tick_params(axis='both', which='major', labelsize=15)
plt.ylim(0.9,1.1)
plt.xlim(0.001,30)
plt.xlabel('k[h/Mpc]',**font1)
plt.legend(loc='best',prop={'size': 14})
plt.tight_layout(pad=3.0)


dat = np.loadtxt("linear/mnu_0.4_nupower_z0.dat")
dat2 = np.loadtxt("linear/mnu_0.4_nupower_z1.dat")

datb = np.loadtxt("linear/nupower_z0.dat")
dat2b = np.loadtxt("linear/nupower_z1.dat")

ax = plt.subplot(1, 3, 3)
plt.xscale('log')
plt.plot(dat[:,0], dat[:,1]/datb[:,1] ,linestyle='dashed', color='b', label= 'z=0',alpha=1.)
plt.plot(dat2[:,0], dat2[:,1]/dat2b[:,1] ,linestyle='dashed', color='b', label= 'z=0',alpha=1.)
ax.tick_params(axis='both', which='major', labelsize=15)
plt.ylim(0.9,1.1)
plt.xlim(0.001,30)
plt.xlabel('k[h/Mpc]',**font1)
plt.legend(loc='best',prop={'size': 14})
plt.tight_layout(pad=3.0)


plt.savefig('plots/linear.png', format="png", bbox_inches = 'tight')
```
